## Exercise 2: list methods

In this exercise, we will focus on the following list methods:

a.) append 
        append has one positional parameter (object)
        the goal of this method is to add an item to a list, and it returns None.
        e.g.
        letters = ["a", "b", "c"]
        letters.append(d)          #this will add d to the list
                    

b.) count
        count also has one positional parameter (value)
        the goal of this method is to count how often does a value appear in a list, and it returns           an integer (the amount of times it appeared)
        e.g.
        letters = ["a", "b", "c", "a"]
        letters.count(a)           #this count the amount of times that a is present in 'letters'

c.) index
        index has one positional parameter (value) but also 2 keyword parameters (start and stop).
        it is used to locate the first index of a value, and it returns an integer.
        e.g.
        friend_list = ['John', 'Bob', 'John', 'Marry', 'Bob']
        friend_list.index('Bob')   #returns 1, the index of the first Bob,
        friend_list.index('John', 1)  #starting from location 1, it finds the first index for John,
                                        therefore it being 2, and not 0
        
d.) insert
        insert has one positional parameter (index) and one keyword parameter (element).
        it is used to add a value to the list in a specific index, and it returns None.
        e.g.
        letters = ["a", "b", "c", "a"]
        letters.insert(1, "x")   #it inserts "x" into the list letters at index 1, resulting in
        the following new list:
            letters = ["a", "x", "b", "c", "a"]
        

e.) pop
        pop only has one positional parameter (object).
        It's function is to remove a value from in a specific index of in a list. when object is 
        not specified, it will remove the last element of the list. It returns the removed value.
        e.g.
        letters = ["a", "b", "c", "a"]
        letters.pop()   #removes and return the last "a" from the list
        e.g.
        letters = ["a", "b", "c", "a"]
        letters.pop(1)   #removes and return "b", as it is in index 1 in the list

#### Exercise 6d: preprocessing text in a more general way
Please write a function that replaces all characters that the user provides by empty spaces.

The function is:
* is called `clean_text_general`
* has one positional parameter `text`
* has one keyword parameter `chars_to_remove`, which is a set (set the default to {'\n', ',', '.', '"'})
* return a string, e.g., the cleaned text

When the user provides a different value to `chars_to_remove`, e.g., {'a'}, then only those characters should be replaced by empty spaces in the text.

```python
def clean_text_general(text, chars_to_remove={'\n', ',', '.', '"'}):
    new_text = text
    for char in chars_to_remove:
        new_text = new_text.replace(char, " ")
    return str(new_text)
```

#### Exercise 6e: including and excluding words
Please write Exercise 5a as a function. 

The function:
* is called `exclude_and_count`
* has one positional parameter `words`, which is a list of strings.
* creates a dictionary in which the **key** is a word and the **value** is the frequency of that word. 
* words are excluded if they meet one of the following criteria:
    * end with the letter e
    * start with the letter t
    * start with the letter c and end with the letter w (both conditions must be met)
    * have six or more letters
* returns a dictionary in which the **key** is the word and the **value** is the frequency of the word. 

```python
def exclude_and_count(words):
    new_words = []
    for word in words:
        if word[0] == 't':
            continue
        if word[-1] == 'e':
            continue
        if word[0] == 'c' and word[-1] == 'w':
            continue
        if len(word) >= 6:
            continue
        else:
            new_words.append(word)
     
    word_dictionary = {}
    for word in new_words:
        if word in word_dictionary:
            word_dictionary[word] +=1
        else:
            word_dictionary[word] = 1
    return word_dictionary

```
